```python
import zipfile  

# Path to the ZIP file
zip_path = "D:\E\Github\Whatsapp_Chat_Analysis\WhatsApp Chat with Varunnnn.zip"

# Destination folder to extract the file
destination_folder = "D:\E\Github\Whatsapp_Chat_Analysis"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    # Get the name of the first file in the ZIP
    text_file = zip_ref.namelist()[0]  # Assuming only one file inside

    # Extract the text file
    zip_ref.extract(text_file, destination_folder)
    print(f"Extracted: {text_file} to {destination_folder}")
```

```python
import re 
import pandas as pd 
```

```python
f = open("D:\E\Github\Whatsapp_Chat_Analysis\WhatsApp Chat with Varunnnn.txt",'r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
print(data)
```

```python
pattern = '\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}\s?[ap]m -'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
dates = [date.replace("\u202f", " ") for date in dates]
dates
```

```python
df = pd.DataFrame({'User_message':messages, 'Dates':dates})
df.head()
```

```python
df['Dates'] = pd.to_datetime(df['Dates'],format="%d/%m/%y, %I:%M %p -")
df['Dates']
```

```python
df.head()
```

```python
user = []
message = []
for messages in df['User_message']:
    entry = re.split('([\w\W]+?):\s',messages)
    if entry[1:]: #user entry 
        user.append(entry[1])
        message.append(" ".join(entry[2:]))
    else:
        user.append('group notification')
        message.append(entry[0])
```

```python
user
```

```python
message
```

```python
df['User'] = user
df['Message'] = message
df.drop(columns=['User_message'],inplace=True)
```

```python
df.head()
```

```python
df['date_only'] = df['Dates'].dt.date
df['year'] = df['Dates'].dt.year
df['month_name'] = df['Dates'].dt.month_name()
df['month'] = df['Dates'].dt.month
df['day'] = df['Dates'].dt.day
df['hour'] = df['Dates'].dt.hour
df['minute'] = df['Dates'].dt.minute
df['day_name']= df['Dates'].dt.day_name()
```

#### Maximum Used Words 

```python
# removing media ommitted 
#remove group notifications 
# remove stop words 
```

```python
temp = df[df['User']!='group notification']
temp = temp[temp['Message']!='<Media omitted>\n']
```

```python
temp.head()
```

```python
f = open('stop_hinglish.txt')
stop_words = f.read()
```

```python
words = []
for message in temp['Message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
from collections import Counter
pd.DataFrame(Counter(words).most_common(20))
```
