```python
import re 
import pandas as pd 
```

```python
f = open("D:\E\Github\Whatsapp_Chat_Analysis\\file\WhatsApp Chat with Varunnnn.txt",'r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
print(data)
```

```python
pattern = '\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}\s?[ap]m -'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
dates = [date.replace("\u202f", " ") for date in dates]
dates
```

```python
df = pd.DataFrame({'User_message':messages, 'Dates':dates})
df.head()
```

```python
df['Dates'] = pd.to_datetime(df['Dates'],format="%d/%m/%y, %I:%M %p -")
df['Dates']
```

```python
df.head()
```

```python
user = []
message = []
for messages in df['User_message']:
    entry = re.split('([\w\W]+?):\s',messages)
    if entry[1:]: #user entry 
        user.append(entry[1])
        message.append(" ".join(entry[2:]))
    else:
        user.append('group notification')
        message.append(entry[0])
```

```python
user
```

```python
message
```

```python
df['User'] = user
df['Message'] = message
df.drop(columns=['User_message'],inplace=True)
```

```python
df.head()
```

```python
df['date_only'] = df['Dates'].dt.date
df['year'] = df['Dates'].dt.year
df['month_name'] = df['Dates'].dt.month_name()
df['month'] = df['Dates'].dt.month
df['day'] = df['Dates'].dt.day
df['hour'] = df['Dates'].dt.hour
df['minute'] = df['Dates'].dt.minute
df['day_name']= df['Dates'].dt.day_name()
```

```python
f = open('stop_hinglish.txt')
stop_words = f.read()
```

```python
words = []
for message in temp['Message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
from collections import Counter
pd.DataFrame(Counter(words).most_common(20))
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour + 1))
    else:
        period.append(str(hour) + "-" + str(hour + 1))

df['period'] = period
```

```python
user_heatmap = df.pivot_table(index='day_name', columns='period', values='Message', aggfunc='count').fillna(0)
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(user_heatmap)
plt.xticks(rotation=0)
plt.show()
```
